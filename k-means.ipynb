```python
clusters = [] #[KMeans(n_clusters=i, random_state=0).fit(trX)  for i in range(1,5,1)]
for i in range(1,60,2):
    print(i)
    clusters.append(KMeans(n_clusters=i, random_state=0).fit(trX))
```

```python
plt.figure(figsize=(16,16))
plt.title("K-Means Display of Elbow Point")
plt.xlabel("Cluster Size")
plt.ylabel("Energy")
plt.plot([len(cluster.cluster_centers_) for cluster in clusters],[-cluster.score(trX) for cluster in clusters ])
plt.show()
```
