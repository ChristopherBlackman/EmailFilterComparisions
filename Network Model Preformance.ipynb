```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#new model layout
#model 1 -- minmax   layer 7
#model 2 -- minmax   layer 15 
#model 3 -- minmax   layer 30
#model 4 -- minmax   layer 60
#model 5 -- pca 1500 layer 15
#model 6 -- pca 1500 layer 30
#model 7 -- naive bayes
#model 8 -- random forest
models = [pd.read_csv("output/model_{0}.csv".format(i)) for i in range(1,9)]
times  = [1000.0*np.average(model['time_per_iteration(s)'].tolist()) for model in models]
labels = ["model {0}".format(i) for i in range(1,9)] 
labels = ["7-layer-minmax","15-layer-minmax","30-layer-minmax","60-layer-minmax","15-layer-pca","30-layer-pca","naive bayes","random forest"]
```

```python
x = np.arange(len(times))
fig = plt.figure()
plt.figure(figsize=(16,16))
plt.title("Average Preformance of Models Per Prediction")
plt.xlabel("Models")
plt.ylabel("Timer Per Prediction (Milliseconds)")
plt.bar(x, times,tick_label=labels, color='b')
plt.savefig('figures/preformance_models.png')
```

```python
models = [pd.read_csv("output/model_{0}.csv".format(i)) for i in range(1,7)]
accuracy = [model["accuracy"].tolist() for model in models]
```

```python
fig = plt.figure()
plt.figure(figsize=(16,16))
plt.title("Validation Accuracy of Nueral Networs While Training")
plt.xlabel("Epocs")
plt.ylabel("Accuracy")
for i in range(len(accuracy)):
    plt.plot(accuracy[i],label=labels[i])
plt.legend()
plt.savefig('figures/training_accuracy.png')
```

```python
fig = plt.figure()
plt.figure(figsize=(16,16))
plt.title("Learning Rate Step Function")
plt.xlabel("Epocs")
plt.ylabel("Learning Rate")
plt.plot(models[0]['learning_rate'].tolist())
plt.savefig('figures/step_function.png')
```

```python
models = [pd.read_csv("output/model_{0}.csv".format(i)) for i in range(1,9)]
accuracy = [max(model["accuracy"].tolist()) for model in models]
```

```python
x = np.arange(len(accuracy))
fig = plt.figure()
plt.figure(figsize=(16,16))
plt.title("Maximum Average Validation Accuracy Acheived by Models")
plt.xlabel("Models")
plt.ylim((0.85,1.0))
plt.ylabel("Accuracy")
plt.bar(x, accuracy,tick_label=labels, color='b')
plt.savefig('figures/accuracy_max.png')
```
